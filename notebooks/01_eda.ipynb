```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/raw/fbref_PL_2024-25.csv", encoding = "latin-1")
print(df.shape)
print(df.dtypes)
print(df.isnull().sum())
```

```python
print(df['Pos'].value_counts())
print(df['Squad'].value_counts())
```

```python
df_filtred = df[df['Min']>1000] 
print(f"Players kept : {len(df_filtred)}/{len(df)}")
```

```python
print(df_filtred.columns.tolist())
```

```python
columns_to_keep = [
    'Player', 'Pos', 'Squad',
    'Gls', 'Ast', 'G+A', 'xG', 'xAG',
    'PrgC', 'PrgP', 'PrgR', 'Gls.1', 'Ast.1', 'G+A.1', 'G-PK.1',
    'G+A-PK', 'xG.1', 'xAG.1', 'xG+xAG', 'npxG.1', 'npxG+xAG.1'
]

df_clean = df_filtred[columns_to_keep]
print(df.shape)
print(df_clean.isnull().sum())
```

```python
top10 = df_clean.nlargest(10, 'G+A')[['Player', 'Squad', 'G+A']]
print(top10)
```

```python
plt.figure(figsize = (10,6))
plt.barh(top10['Player'], top10['G+A'], color = 'steelblue')
plt.xlabel('Goals + Assists')
```
